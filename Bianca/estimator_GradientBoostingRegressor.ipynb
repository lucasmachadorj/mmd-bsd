```python
import datetime
import numpy as np
import pandas as pd
import itertools
import matplotlib.pyplot as plt

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.grid_search import GridSearchCV

def SLE(a,b):
    return ((a-b)**2)

def RMSE(a,b):
    return np.sqrt(SLE(a,b).mean())

def fixNegatives(vector):
    for i in range(len(vector)):
        if vector[i]<0:
            vector[i]=0
    pass

%matplotlib inline

datasets = "../datasets/"
train_data = pd.read_csv(datasets+"super_train.csv",sep=',',parse_dates=True,index_col='datetime')
test_data = pd.read_csv(datasets+"super_test.csv",sep=',',parse_dates=True,index_col='datetime')
```

# Treinamento em escala logarítmica

### Preparando treinamento

```python
notfeature=['casual','registered','count']
features = [L for L in train_data.columns if L not in notfeature]
trainday = 13

X_train = train_data[features][train_data.index.day<trainday].values
ylog_train_casual = np.log(train_data['casual'][train_data.index.day<trainday].values +1)
ylog_train_registered = np.log(train_data['registered'][train_data.index.day<trainday].values +1)

X_test = train_data[features][train_data.index.day>=trainday].values
ylog_test_casual = np.log(train_data['casual'][train_data.index.day>=trainday].values +1)
ylog_test_registered = np.log(train_data['registered'][train_data.index.day>=trainday].values +1)

ylog_test_count = np.log(train_data['count'][train_data.index.day>=trainday].values+1)
```

```python
notfeature=['casual','registered','count']
features = [L for L in train_data.columns if L not in notfeature]
trainday = 13

X_train = []
ylog_train_casual = []
ylog_train_registered = []

X_test = []
ylog_test_casual = []
ylog_test_registered = []

ylog_test_count = []

for i in range(6):
    X_train.append(train_data[features][(train_data.index.day<trainday) & (train_data.periodo==i)].values)
    ylog_train_casual.append(np.log(train_data['casual'][(train_data.index.day<trainday) & (train_data.periodo==i)].values +1))
    ylog_train_registered.append(np.log(train_data['registered'][(train_data.index.day<trainday) & (train_data.periodo==i)].values +1))

    X_test.append(train_data[features][(train_data.index.day>=trainday) & (train_data.periodo==i)].values)
    ylog_test_casual.append(np.log(train_data['casual'][(train_data.index.day>=trainday) & (train_data.periodo==i)].values +1))
    ylog_test_registered.append(np.log(train_data['registered'][(train_data.index.day>=trainday) & (train_data.periodo==i)].values +1))

    ylog_test_count.append(np.log(train_data['count'][(train_data.index.day>=trainday) & (train_data.periodo==i)].values+1))

ylog_test_casual_all = np.log(train_data['casual'][(train_data.index.day>=trainday)].values +1)
ylog_test_registered_all = np.log(train_data['registered'][(train_data.index.day>=trainday)].values +1)
```

# Predição

```python
c_params = {'n_estimators':1000,'learning_rate':0.02,'max_depth':4,'max_features':0.5,'min_samples_leaf':5}
ylog_pred_casual = []

gbrlog_casual = GradientBoostingRegressor(**c_params)
for i in range(6):
    gbrlog_casual.fit(X_train[i],ylog_train_casual[i])
    ylog_pred_casual.append(gbrlog_casual.predict(X_test[i]))
```

```python
for i in range(6):
    fixNegatives(ylog_pred_casual[i])
    print "ERROR %f" % RMSE(ylog_test_casual[i],ylog_pred_casual[i])
```

```python
df_casual = []
for i in range(6):
    df_casual.append(train_data[['periodo']][(train_data.index.day>=trainday) & (train_data.periodo==i)])
    df_casual[i]['pred'] = np.array(ylog_pred_casual[i])

frames = [df_casual[0],df_casual[1],df_casual[2],df_casual[3],df_casual[4],df_casual[5]]
all_periods_casual = (pd.concat(frames)).sort()

print "ERROR %f" % RMSE(ylog_test_casual_all,all_periods_casual['pred'])
```

```python
casual_frame = pd.DataFrame(all_periods_casual['pred'],columns=['pred'])
casual_frame['datetime'] = train_data.index[train_data.index.day>=trainday]
casual_frame.set_index('datetime',inplace=True)
casual_frame['test'] = all_periods_casual['pred']
casual_frame['sle'] = SLE(all_periods_casual['pred'],ylog_test_casual_all)
casual_frame['event'] = train_data[train_data.index.day>=trainday]['event']

print casual_frame[casual_frame['sle']>3]

casual_frame[['test','pred']].plot(figsize=(20,5))
plt.show()
casual_frame['sle'].plot(figsize=(20,5))
plt.show()
```

# Preparando Submissão: escala logarítmica venceu

```python
notfeature=['casual','registered','count']
features = [L for L in train_data.columns if L not in notfeature]

X_train = []
ylog_train_casual = []
ylog_train_registered = []

X_test = []

for i in range(7):
    X_train.append(train_data[features][(train_data.periodo==i)].values)
    ylog_train_casual.append(np.log(train_data['casual'][(train_data.periodo==i)].values +1))
    ylog_train_registered.append(np.log(train_data['registered'][(train_data.periodo==i)].values +1))

    X_test.append(test_data[features][(test_data.periodo==i)].values)
```

```python
c1 = {'n_estimators':2000,'learning_rate': 0.02,'max_depth': 4,'max_features': 0.3,'min_samples_leaf': 17}
c2 = {'n_estimators':1000,'learning_rate':0.02,'max_depth':4,'max_features':0.5,'min_samples_leaf':5}

gbrlog_casual = GradientBoostingRegressor(**c2)

ylog_pred_casual = []
for i in range(7):
    gbrlog_casual.fit(X_train[i],ylog_train_casual[i])
    ylog_pred_casual.append(gbrlog_casual.predict(X_test[i]))
    fixNegatives(ylog_pred_casual[i])
```

```python
df_casual = []
for i in range(7):
    df_casual.append(test_data[['periodo']][(test_data.periodo==i)])
    df_casual[i]['pred'] = np.array(ylog_pred_casual[i])

frames = [df_casual[0],df_casual[1],df_casual[2],df_casual[3],df_casual[4],df_casual[5],df_casual[6]]
ylog_pred_casual = (pd.concat(frames)).sort()['pred']
```

```python
r1 = {'n_estimators':2000,'learning_rate': 0.1,'max_depth': 4,'max_features': 0.3,'min_samples_leaf': 17}
r2 = {'n_estimators':1000,'learning_rate': 0.05,'max_depth': 4,'max_features': 0.3,'min_samples_leaf': 5}

gbrlog_registered = GradientBoostingRegressor(**r2)

ylog_pred_registered = []
for i in range(7):
    gbrlog_registered.fit(X_train[i],ylog_train_registered[i])
    ylog_pred_registered.append(gbrlog_registered.predict(X_test[i]))
    fixNegatives(ylog_pred_registered[i])
```

```python
df_registered = []
for i in range(7):
    df_registered.append(test_data[['periodo']][(test_data.periodo==i)])
    df_registered[i]['pred'] = np.array(ylog_pred_registered[i])

frames = [df_registered[0],df_registered[1],df_registered[2],df_registered[3],df_registered[4],df_registered[5],df_registered[6]]
ylog_pred_registered = (pd.concat(frames)).sort()['pred']
```

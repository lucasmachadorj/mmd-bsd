```python
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.ensemble import GradientBoostingRegressor

def fixNegatives(vector):
    for i in range(len(vector)):
        if vector[i]<0:
            vector[i]=0
    pass

def SLE(p,a):
    return np.power(np.log(np.array(p)+1) - np.log(np.array(a)+1), 2)

def RMSLE(p,a):
    return np.sqrt(np.mean(SLE(p,a)))

%matplotlib inline

datasets = "datasets/"
```

# Engenharia de Features

```python
train_data = pd.read_csv(datasets+"train.csv",sep=',')
train_data['datetime']  = pd.to_datetime(train_data['datetime'],infer_datetime_format=True)

train_data['year'] = train_data.apply(lambda row: pd.to_datetime(row['datetime']).year,axis=1)
train_data['month'] = train_data.apply(lambda row: pd.to_datetime(row['datetime']).month,axis=1)
train_data['day'] = train_data.apply(lambda row: pd.to_datetime(row['datetime']).day,axis=1)
train_data['hour'] = train_data.apply(lambda row: pd.to_datetime(row['datetime']).hour,axis=1)
train_data['weekday'] = train_data.apply(lambda row: pd.to_datetime(row['datetime']).dayofweek,axis=1)

train_data = train_data.set_index('datetime')
train_data.head()
```

```python
daytemps = train_data['temp'].groupby([train_data.index.year,train_data.index.month,train_data.index.day])
meant = daytemps.mean()
maxt = daytemps.max()
mint = daytemps.min()


train_data['max temp'] = train_data.apply(lambda row: maxt[(row['year'],row['month'],row['day'])], axis=1)
train_data['avg temp'] = train_data.apply(lambda row: meant[(row['year'],row['month'],row['day'])], axis=1)
train_data['min temp'] = train_data.apply(lambda row: mint[(row['year'],row['month'],row['day'])], axis=1)

train_data
```

## Treinamento em escala natural

```python
def getcol(matrix,c):
    return np.array([line[c] for line in matrix])

notfeature = ['casual','registered','count']
features = [L for L in train_data.columns if L not in notfeature]
trainday = 16

X_train = train_data[features][train_data['day']<trainday].values
# For multi-dimensional regression 
y_train = train_data[['casual','registered']][train_data['day']<trainday].values
# For unidimensional regression
y_train_casual = getcol(y_train,0)
y_train_registered = getcol(y_train,1)

X_test = train_data[features][train_data['day']>=trainday].values
# For multi-dimensional regression
y_test = train_data[['casual','registered']][train_data['day']>=trainday].values
# For unidimensional regression
y_test_casual = getcol(y_test,0)
y_test_registered = getcol(y_test,1)


y_true = train_data[['count']][train_data['day']>=trainday].values
y_true = getcol(y_true,0)
```

```python
setup = {'n_estimators':100}

gbr_casual = GradientBoostingRegressor(**setup).fit(X_train,y_train_casual)
y_pred_casual = gbr_casual.predict(X_test)
fixNegatives(y_pred_casual)

fig, ax = plt.subplots(figsize=(20,5))
plt.plot(y_test_casual,label='casual')
plt.plot(y_pred_casual,label='prediction')
plt.legend()

print "CASUAL RMSLE: {}".format(RMSLE(y_test_casual,y_pred_casual))


gbr_registered = GradientBoostingRegressor(**setup).fit(X_train,y_train_registered)
y_pred_registered = gbr_registered.predict(X_test)
fixNegatives(y_pred_registered)

fig, ax = plt.subplots(figsize=(20,5))
plt.plot(y_test_registered,label='registered')
plt.plot(y_pred_registered,label='prediction')
plt.legend()

print "CASUAL RMSLE: {}".format(RMSLE(y_test_registered,y_pred_registered))


y_try = y_pred_casual+y_pred_registered

fig, ax = plt.subplots(figsize=(20,5))
plt.plot(y_true,label='count')
plt.plot(y_try,label='prediction')
plt.legend()


print "RMSLE: {}".format(RMSLE(y_true,y_try))
```

## Treinamento em escala logarítmica

```python
delta = 1

ylog_train_casual = np.log(y_train_casual+delta)
ylog_train_registered = np.log(y_train_registered+delta)

ylog_test_casual = np.log(y_test_casual+delta)
ylog_test_registered = np.log(y_test_registered+delta)
```

```python
setup = {'n_estimators':300,'learning_rate':0.2,'verbose':0,'subsample':0.9}

gbrlog_casual = GradientBoostingRegressor(**setup).fit(X_train,ylog_train_casual)
ylog_pred_casual = gbrlog_casual.predict(X_test)
fixNegatives(ylog_pred_casual)

fig, (ax,bx) = plt.subplots(nrows=2,figsize=(20,10))
ax.plot(y_test_casual,label='casual')
ax.plot(np.exp(ylog_pred_casual)-delta,label='prediction')
bx.plot(SLE(y_test_casual,np.exp(ylog_pred_casual)-delta),label='SLE')
bx.set_ylim(top=10)
plt.legend()

print "CASUAL RMSLE: {}".format(RMSLE(y_test_casual,np.exp(ylog_pred_casual)-delta))


gbrlog_registered = GradientBoostingRegressor(**setup).fit(X_train,ylog_train_registered)
ylog_pred_registered = gbrlog_registered.predict(X_test)
fixNegatives(ylog_pred_registered)

fig, (ax,bx) = plt.subplots(nrows=2,figsize=(20,10))
ax.plot(y_test_registered,label='registered')
ax.plot(np.exp(ylog_pred_registered)-delta,label='prediction')
bx.plot(SLE(y_test_registered,np.exp(ylog_pred_registered)-delta),label='SLE')
bx.set_ylim(top=10)
plt.legend()

print "REGISTERED RMSLE: {}".format(RMSLE(y_test_registered,np.exp(ylog_pred_registered)-delta))


y_try = np.array(np.exp(ylog_pred_casual)-delta)+np.array(np.exp(ylog_pred_registered)-delta)

fig, (ax,bx) = plt.subplots(nrows=2,figsize=(20,10))
ax.plot(y_true,label='count')
ax.plot(y_try,label='prediction')
bx.plot(SLE(y_true,y_try),label='SLE')
bx.set_ylim(top=10)
plt.legend()

print "TOTAL RMSLE: {}".format(RMSLE(y_true,y_try))
```

```python
setup = {'n_estimators':500}

gbrlog_casual = GradientBoostingRegressor(**setup).fit(X_train,ylog_train_casual)
ylog_pred_casual = gbrlog_casual.predict(test_data[features])
fixNegatives(ylog_pred_casual)

gbrlog_registered = GradientBoostingRegressor(**setup).fit(X_train,ylog_train_registered)
ylog_pred_registered = gbrlog_registered.predict(test_data[features])
fixNegatives(ylog_pred_registered)

y_try = np.array(np.exp(ylog_pred_casual)-1)+np.array(np.exp(ylog_pred_registered)-1)
```

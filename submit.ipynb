```python
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def fixNegatives(vector):
    for i in range(len(vector)):
        if vector[i]<0:
            vector[i]=0

def SLE(p,a):
    le = np.log(np.array(p)+1) - np.log(np.array(a)+1)
    return np.power(le, 2)

def RMSLE(p,a):
    return np.sqrt(np.mean(SLE(p,a)))

%matplotlib inline

datasets = "datasets/"
```

```python
train_data = pd.read_csv(datasets+"train.csv",sep=',')
train_data['datetime']  = pd.to_datetime(train_data['datetime'],infer_datetime_format=True)

test_data = pd.read_csv(datasets+"test.csv",sep=',')
test_data['datetime']  = pd.to_datetime(test_data['datetime'],infer_datetime_format=True)
```

```python
def dow2str(dayofweek):
    d = {0:'monday',1:'tuesday',2:'wednesday',3:'thursday',4:'friday',5:'saturday',6:'sunday'}
    return d[dayofweek]

train_data['year'] = train_data.apply(lambda row: pd.to_datetime(row['datetime']).year,axis=1)
train_data['month'] = train_data.apply(lambda row: pd.to_datetime(row['datetime']).month,axis=1)
train_data['day'] = train_data.apply(lambda row: pd.to_datetime(row['datetime']).day,axis=1)
train_data['hour'] = train_data.apply(lambda row: pd.to_datetime(row['datetime']).hour,axis=1)
train_data['weekday'] = train_data.apply(lambda row: pd.to_datetime(row['datetime']).dayofweek,axis=1)

test_data['year'] = test_data.apply(lambda row: pd.to_datetime(row['datetime']).year,axis=1)
test_data['month'] = test_data.apply(lambda row: pd.to_datetime(row['datetime']).month,axis=1)
test_data['day'] = test_data.apply(lambda row: pd.to_datetime(row['datetime']).day,axis=1)
test_data['hour'] = test_data.apply(lambda row: pd.to_datetime(row['datetime']).hour,axis=1)
test_data['weekday'] = test_data.apply(lambda row: pd.to_datetime(row['datetime']).dayofweek,axis=1)


train_data = train_data.set_index('datetime')
test_data = test_data.set_index('datetime')
```

```python
def getcol(matrix,c):
    return np.array([line[c] for line in matrix])

features = [L for L in train_data.columns if L not in ['casual','registered','count']]

X_train = train_data[features].values
# For multi-dimensional regression 
y_train = train_data[['casual','registered']].values
# For unidimensional regression
y_train_casual = getcol(y_train,0)
y_train_registered = getcol(y_train,1)
```

```python
from sklearn.tree import DecisionTreeRegressor

dtr_casual = DecisionTreeRegressor().fit(X_train,y_train_casual)
dtr_registered = DecisionTreeRegressor().fit(X_train,y_train_registered)
```

```python
y_pred_casual = dtr_casual.predict(test_data)
y_pred_registered = dtr_registered.predict(test_data)
y_try = np.array(y_pred_casual)+np.array(y_pred_registered)
```

```python
submission = pd.read_csv(datasets+"sampleSubmission.csv",sep=',')
submission['datetime']  = pd.to_datetime(submission['datetime'],infer_datetime_format=True)
submission = submission.set_index('datetime')

submission['count'] = y_try
submission.head()
```

```python
fig, ax = plt.subplots(figsize=(20,5))
train_data['count'].plot(ax=ax)
submission['count'].plot(ax=ax,color='r')
```

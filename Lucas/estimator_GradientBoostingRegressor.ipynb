```python
import datetime
import numpy as np
import pandas as pd
import itertools
import matplotlib.pyplot as plt

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.grid_search import GridSearchCV

def RMSE(a,b):
    return np.sqrt(((a-b)**2).mean())

def fixNegatives(vector):
    for i in range(len(vector)):
        if vector[i]<0:
            vector[i]=0
    pass

%matplotlib inline

datasets = "../datasets/"
train_data = pd.read_csv(datasets+"super_train.csv",sep=',',parse_dates=True,index_col='datetime')
test_data = pd.read_csv(datasets+"super_test.csv",sep=',',parse_dates=True,index_col='datetime')
```

# Treinamento em escala logarítmica

### Preparando treinamento

```python
notfeature=['casual','registered','count']
features = [L for L in train_data.columns if L not in notfeature]
trainday = 13

X_train = train_data[features][train_data.index.day<trainday].values
ylog_train_casual = np.log(train_data['casual'][train_data.index.day<trainday].values +1)
ylog_train_registered = np.log(train_data['registered'][train_data.index.day<trainday].values +1)

X_test = train_data[features][train_data.index.day>=trainday].values
ylog_test_casual = np.log(train_data['casual'][train_data.index.day>=trainday].values +1)
ylog_test_registered = np.log(train_data['registered'][train_data.index.day>=trainday].values +1)

y_test_count = train_data['count'][train_data.index.day>=trainday].values
```

# Grid Search

```python
param_grid = {'learning_rate': [0.1, 0.05, 0.02, 0.01],
              'max_depth': [2, 4, 6],
              'min_samples_leaf': [3, 5, 9, 17],
              'max_features': [1.0, 0.5, 0.3]
              }
est = GradientBoostingRegressor(n_estimators=1000)
```

```python
gs_cv = GridSearchCV(est, param_grid, n_jobs=3, verbose=1).fit(X_train, ylog_train_casual)
gs_cv.best_params_
```

```python
gs_cv = GridSearchCV(est, param_grid, n_jobs=3, verbose=1).fit(X_train, ylog_train_registered)
gs_cv.best_params_
```

```python
c1 = {'n_estimators':500,'learning_rate':0.05,'max_depth':4,'max_features':0.5,'min_samples_leaf':9} # 0.521181
c2 = {'n_estimators':1000,'learning_rate':0.02,'max_depth':4,'max_features':0.5,'min_samples_leaf':5}

gbrlog_casual = GradientBoostingRegressor(**c2)
gbrlog_casual.fit(X_train,ylog_train_casual)
ylog_pred_casual = gbrlog_casual.predict(X_test)
fixNegatives(ylog_pred_casual)

print "ERROR %f" % RMSE(ylog_test_casual,ylog_pred_casual)
```

```python
r1 = {'n_estimators':500,'learning_rate':0.1,'max_depth':4,'max_features':0.3,'min_samples_leaf':17} # 0.314250
r2 = {'n_estimators':1000,'learning_rate': 0.05,'max_depth': 4,'max_features': 0.3,'min_samples_leaf': 5}

gbrlog_registered = GradientBoostingRegressor(**r2)
gbrlog_registered.fit(X_train,ylog_train_registered)
ylog_pred_registered = gbrlog_registered.predict(X_test)
fixNegatives(ylog_pred_registered)

print "ERROR %f" % RMSE(ylog_test_registered,ylog_pred_registered)
```

# Preparando Submissão: escala logarítmica venceu

```python
notfeature=['casual','registered','count']
features = [L for L in train_data.columns if L not in notfeature]

X_train = train_data[features].values
ylog_train_casual = np.log(train_data['casual'].values +1)
ylog_train_registered = np.log(train_data['registered'].values +1)

X_test = test_data[features].values
```

```python
c1 = {'n_estimators':500,'learning_rate':0.05,'max_depth':4,'max_features':0.5,'min_samples_leaf':9} # 0.521181
c2 = {'n_estimators':1000,'learning_rate':0.02,'max_depth':4,'max_features':0.5,'min_samples_leaf':5}

gbrlog_casual = GradientBoostingRegressor(**c2)
gbrlog_casual.fit(X_train,ylog_train_casual)
ylog_pred_casual = gbrlog_casual.predict(X_test)
fixNegatives(ylog_pred_casual)
```

```python
r1 = {'n_estimators':500,'learning_rate':0.1,'max_depth':4,'max_features':0.3,'min_samples_leaf':17} # 0.314250
r2 = {'n_estimators':1000,'learning_rate': 0.05,'max_depth': 4,'max_features': 0.3,'min_samples_leaf': 5}

gbrlog_registered = GradientBoostingRegressor(**r2)
gbrlog_registered.fit(X_train,ylog_train_registered)
ylog_pred_registered = gbrlog_registered.predict(X_test)
fixNegatives(ylog_pred_registered)
```

```python
submission = pd.read_csv(datasets+"sampleSubmission.csv",sep=',',parse_dates=True,index_col='datetime')

submission['count'] = np.exp(ylog_pred_casual)-1 + np.exp(ylog_pred_registered)-1
```

```python
fig, ax = plt.subplots(figsize=(20,5))
train_data['count'].plot(ax=ax)
submission['count'].plot(ax=ax,color='r')
```
